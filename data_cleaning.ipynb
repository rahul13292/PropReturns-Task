```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('mumbai_realestate.csv')
```

```python
df
```

```python
#removing duplicate entries
df2=df.drop_duplicates()  
```

```python
#drop columns columns which are irrelevant
df2=df2.drop(columns='List no. 2')
```

```python
#convert to datetime data type
df2['Year'] = pd.to_datetime(df2['Year'])
```

```python
df2
```

```python
# Apply string slicing
df2['Other information'] = df2['Other information'].fillna('').apply(lambda x: x[2:]) 
```

```python
#renaming columns according to mapping
column_rename_mapping = {
    'Sl no.': 's_no',
    'Diarrhea no.': 'doc_no', 
    'diarrhea type': 'doc_type',
    'Du. Prohibit. Office': 'sro_name',  #sro --> sub-registrar office
    'Will write': 'Buyer_name',
    'Will write down': 'Seller_name'
}

df2.rename(columns=column_rename_mapping, inplace=True)
```

```python
df2
```

```python
#checking which columns have NaN values
nan_columns = df2.columns[df2.isna().any()].tolist()
print(nan_columns)
```

```python
#count of NaN values in column
nan_count = df2['Seller_name'].isna().sum()
print(nan_count)
```

```python
#replace empty cells with NaN values
df2 = df2.replace(r'^\s*$', np.nan, regex=True)
```

```python
#convert datatype in particular columns to string in order to extract names into a dict
columns_to_convert1 = ['doc_type','sro_name','Year','Buyer_name','Seller_name','Other information','Link']
df2[columns_to_convert1] = df2[columns_to_convert1].astype(str)
#convert other columns to relevant data types
columns_to_convert2 = ['s_no','doc_no']
df2[columns_to_convert2] = df2[columns_to_convert2].astype(int)
```
